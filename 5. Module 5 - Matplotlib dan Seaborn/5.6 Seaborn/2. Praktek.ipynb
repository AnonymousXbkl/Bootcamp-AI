# Praktek Seaborn

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

### Membuat diagram sederhana dengan Seaborn

```python
data = sns.load_dataset('iris')
data
```

### 

### Membuat linechart di Seaborn

```python
data = sns.load_dataset('iris')
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
```

### Membuat diagram sederhana dengan Seaborn dan PyPlot

```python
plt.figure(figsize=(10, 12))

plt.subplot(211)
sns.set_palette('vlag')
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Title using Matplotlib Function')
plt.plot()


plt.subplot(212)
sns.set_palette('Accent')
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Title using Matplotlib Function')
plt.plot()

plt.show()

```

### Membuat Scatter Plot di Seaborn


```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
```

### Membuat Barchart menggunakan Seaborn


```python
plt.figure(figsize=(4, 4))

sns.barplot(data=data, x='species', y='sepal_width', width=0.5)
```

### Menghitung jumlah variable dengan plot

```python


sns.countplot(data=data, x='species')
plt.show()
```

### Membuat Histogram dengan Seaborn

```python
sns.histplot(data=data, x='sepal_width', hue='sepal_length', palette=sns.color_palette("flare", as_cmap=True))

plt.show()
```

### Membuat Box Plot

```python
sns.boxplot(data=data, x='species', y='sepal_width', hue='species', palette=sns.color_palette('deep', 3))
plt.show()
```

### Membuat Violin Plot

```python
sns.violinplot(data=data, x='species', y='sepal_width', hue='species', palette=sns.color_palette('pastel', 3))
plt.show()
```

```python
sns.color_palette()
```

### Membuat Pair Plot

```python
sns.pairplot(data=data, hue='species')
```

### Membuat Heatmap

```python
# Menghitung korelasi hanya untuk kolom numerik
tc = data.select_dtypes(include='number').corr()

# Membuat heatmap
sns.heatmap(tc, annot=True, cmap="flare")
plt.show()
```
