# Praktek Subplot

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat 2 plot dengan 1 visualisasi

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints)

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([6, 10, 4, 13])
plt.plot(xpoints, ypoints)

plt.show()
```

### Membuat 2 plot dengan 2 visualisasi

```python
# Plot 1
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints)

plt.show()

# Plot 2
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([6, 10, 4, 13])
plt.plot(xpoints, ypoints)

plt.show()
```

### Membuat 2 plot dengan 2 visualisasi secara horizontal

```python
plt.figure(figsize=(10, 4))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(1, 2, 1)    # 1 Row 2 Column
plt.plot(x, y, marker= 'o', mfc='green', ms= 4)
plt.grid(axis='y', linestyle='--')


#Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(1, 2, 2)     # 1 Row 2 Column
plt.plot(x, y, color='red', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')
plt.show()


```

### Membuat 2 plot dengan 2 visualisasi secara vertikal

```python
plt.figure(figsize=(5, 6))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(211)    # 2 Row 1 Column
plt.plot(x, y, marker= 'o', mfc='green', ms= 4)
plt.grid(axis='y', linestyle='--')


#Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(212)     # 2 Row 1 Column
plt.plot(x, y, color='red', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')
plt.show()
```

### Membuat banyak plot

```python
plt.figure(figsize=(10, 6))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(2, 3, 1)    # 2 Row 3 Column
plt.plot(x, y, marker= 'o', mfc='green', ms= 4)
plt.grid(axis='y', linestyle='--')


#Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(2, 3, 2)     # 2 Row 3 Column
plt.plot(x, y, color='red', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')


#Plot 3
x = np.array([0, 1, 2, 3])
y = np.array([20, 0, 120, 10])

plt.subplot(2, 3, 3)     # 2 Row 3 Column
plt.plot(x, y, color='purple', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')


#Plot 4
x = np.array([0, 1, 2, 3])
y = np.array([20, 10, 40, 80])

plt.subplot(2, 3, 4)     # 2 Row 3 Column
plt.plot(x, y, color='darkgrey', marker='*', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')


#Plot 5
x = np.array([0, 1, 2, 3])
y = np.array([100, 30, 80, 10])

plt.subplot(2, 3, 5)     # 2 Row 3 Column
plt.plot(x, y, color='orange', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')


#Plot 5
x = np.array([0, 1, 2, 3])
y = np.array([20, 170, 40, 120])

plt.subplot(2, 3, 6)     # 2 Row 3 Column
plt.plot(x, y, color='black', marker='o', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')


plt.show()
```

### Memberi title pada 2 plot

```python
plt.figure(figsize=(10, 4))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(1, 2, 1)    # 1 Row 2 Column
plt.plot(x, y, marker= 'o', mfc='green', ms= 4)
plt.grid(axis='y', linestyle='--')
plt.title('SALES')    # Menggunakan function plt.title()


#Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(1, 2, 2)     # 1 Row 2 Column
plt.plot(x, y, color='red', marker='D', ms=4, mfc='green', mec='green')
plt.grid(axis='y', linestyle='--')
plt.title('INCOME')     # Menggunakan function plt.title()

plt.show()
```
