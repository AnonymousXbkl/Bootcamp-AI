# Praktek Diagram

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)
plt.show()


# Dari Scatter plot dibawah dapat dilihat semakin tinggi nilai x maka semakin rendah nilai 
```

### Memberi warna pada Scatter

```python
# Scatter 1
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y, color='orange')


# Scatter 2
x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7, 14, 12])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])

plt.scatter(x, y, color='#90ee90')


plt.show()
```

### Mengatur ukuran untuk tiap buble

```python
size = np.array([10, 25, 50, 100, 250, 500, 30, 45, 5, 150, 300, 100, 40])

# Scatter 1
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y, color='orange', s=size)


# Scatter 2
x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7, 14, 12])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])

plt.scatter(x, y, color='darkgrey')


plt.show()
```

### Membuat Bar Chart

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Barchart Horizontal

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height=0.5, color='green')      # zKita bisa tambahkan widht atau height untuk atur ukuran bar
plt.grid(axis='x', linestyle='--')
plt.show()
```

### Menampilkan PyChart

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Manggos']
x = np.array([25, 35, 25, 15])
colors = ['green', 'lightcoral', 'gold', 'cyan']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(x, colors=colors, autopct='%1.1f%%', shadow=True, startangle=180, explode=explode)
plt.show()
```

### Menambahkan Legends pada PieChart

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Manggos']
x = np.array([25, 35, 25, 15])
colors = ['green', 'lightcoral', 'gold', 'cyan']
explode = (0, 0.2, 0, 0.1)

plt.pie(x, colors=colors, 
        autopct='%1.1f%%',
        labels=labels, 
        shadow=True, 
        startangle=180, 
        explode=explode)
plt.legend(loc=(0.9, 0.8))
plt.show()
```
