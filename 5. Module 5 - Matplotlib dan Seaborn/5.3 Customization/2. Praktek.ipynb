# Praktek Customization

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat diagramd dengan marker value

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o') # Bisa menggunakan marker lain seperti '*', 'D', dll
plt.show()
```

### Membuat diagram dengan line value

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, ':')    # By default nilainya '-', bisa kita ganti sesuai preferensi misal jadi '--' maka jadi garis putus2
plt.show()
```

### Membuat diagram dengan custom color value

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'b')    # bisa ganti ke warna lain misal 'g' (green), 'r' (red)
plt.show()
```

### Membuat diagram dengan format marker, line, color

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o', linestyle='--', color='r')
# bisa juga disingkat jadi = plt.plot(xpoints, ypoints, 'o--r)
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'D:b')
plt.show()
```

### Membuat diagram dengan marker parameter

```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o', ms=6, mec='green', mfc='#4cafff')
# ms= besar marker
# mec= warna outline marker
# mfc= warna fill marker
plt.show()
```

### Membuat diagram dengan tittle dan label

```python
font_title = {'family': 'Arial', 'color': 'darkgreen', 'size': '20', 'weight': 'bold'}
font_label = {'family': 'serif', 'color': 'darkgrey', 'size': '10'}


xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o', ms=6, mec='green', mfc='#4cafff')
plt.title('Sport Watch Data', fontdict=font_title)
plt.xlabel('Avg Pulse', fontdict=font_label)
plt.ylabel('Calories Burnage', fontdict=font_label)
plt.show()
```

### Memberi grid di plot

```python
font_title = {'family': 'Arial', 'color': 'darkgreen', 'size': '20', 'weight': 'bold'}
font_label = {'family': 'serif', 'color': 'darkgrey', 'size': '10'}


xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o', ms=6, mec='green', mfc='#4cafff')
plt.title('Sport Watch Data', fontdict=font_title)
plt.xlabel('Avg Pulse', fontdict=font_label)
plt.ylabel('Calories Burnage', fontdict=font_label)
plt.grid(linestyle='--')
# plt.grid(axis='y')
# plt.grid(axis='x')
plt.show()
```

### Mengubah ukuran diagram

```python
plt.figure(figsize=(10, 4))
font_title = {'family': 'Arial', 'color': 'darkgreen', 'size': '20', 'weight': 'bold'}
font_label = {'family': 'serif', 'color': 'darkgrey', 'size': '10'}


xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o', ms=6, mec='green', mfc='#4cafff')
plt.title('Sport Watch Data', fontdict=font_title)
plt.xlabel('Avg Pulse', fontdict=font_label)
plt.ylabel('Calories Burnage', fontdict=font_label)
plt.grid(axis= 'y', linestyle='--')
plt.show()
```
