# Praktek Matplorlib Plotting

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([0, 20])
ypoints = np.array([180, 0])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat diagram sederhana dengan marker value

```python
xpoints = np.array([0, 20])
ypoints = np.array([0, 180])

plt.plot(xpoints, ypoints, marker='o')
```

### Membuat diagram banyak titik dengan PyPlot


```python
xpoints = np.array([1, 2, 6, 9])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='o')
plt.show()
```
