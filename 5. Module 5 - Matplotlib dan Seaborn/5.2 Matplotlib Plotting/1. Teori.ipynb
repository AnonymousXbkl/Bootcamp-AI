# Plotting

### Matplotlib Plotting
    Matplotlib Plotting memiliki modul khusus bernama pylot yang memuat ragam fungsi untuk kebutuhan visualisasi data dengan diagram seperti Line Chart, Bar Chart, Histogram, Pie Chart dan lain sebagainya.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([0, 20])
ypoints = np.array([5, 180])
zpoints = np.array([20, 125])

plt.plot(xpoints, ypoints, zpoints)
plt.show()
```
