    Kita bisa menggunakan inplace=True untuk menimpa data asli dengan data baru yang sudah kita modifikasi

```python
# Mengatai Missing Value dengan nilai baru

df = pd.read_csv('./dirty_data.csv')
df.fillna('ini nilai yang diganti', inplace=True)   # Inplace True berarti kita mengganti data asli dengan data baru
df
```

```python
# Mengatai Missing Value dengan nilai baru tapi hanya di satu kolom
# Misal kita hanya mengganti di kolom calories

df = pd.read_csv('./dirty_data.csv')
df['Calories'].fillna('Ini nilai yang diganti', inplace=True)
df.tail(15)
```

#### Mengatasi Missing Values dengan nilai Mean

```python
df = pd.read_csv('./dirty_data.csv')
x = df['Calories'].mean()

df['Calories'].fillna(x, inplace=True)
df
```

#### Mengatasi Missing Values dengan nilai Median

```python
df = pd.read_csv('./dirty_data.csv')
x = df['Calories'].median()

df['Calories'].fillna(x, inplace=True)
df
```

#### Mengatasi Missing Values dengan nilai Mode(Modus)

```python
df = pd.read_csv('./dirty_data.csv')
x = df['Calories'].mode()[0]    # Ditambah koma 0

df['Calories'].fillna(x, inplace=True)
df
```

### Menghilangkan duplikat pada DataFrame

```python
df = pd.read_csv('./dirty_data.csv')
df.drop_duplicates(inplace=True)
df
```

### Mengatasi wrong format dengan ubah format

```python
df = pd.read_csv('./dirty_data.csv')
df['Date'] = pd.to_datetime(df['Date'],format='mixed' ,errors='coerce')

# Kita tambahkan format='mixed' karena kita menggunakan berbagai format di dalam kolom itu

df
```

```python
# Coba ubah format 'Calories' jadi to_numeric

df = pd.read_csv('./dirty_data.csv')
df['Calories'] = pd.to_numeric(df['Calories'])
df
```

### Mengatasi Wrong Format dengan Drop Data

```python
df.dropna(subset=['Date'], inplace=True)

df
```

```python
# Drop data menggunakan filter

df = pd.read_csv('./dirty_data.csv')
for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.drop(x, inplace=True)

df
```
